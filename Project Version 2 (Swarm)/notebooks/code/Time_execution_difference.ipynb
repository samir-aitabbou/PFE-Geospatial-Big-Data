# 1 machine 

```python

import time
import matplotlib.pyplot as plt

def perform_task(k):
    
    # Simulating a computationally intensive task
    result = 0
    for i in range(10**k):
        result += i
    time.sleep(k) # Simulating a task that takes 5 seconds to complete
    return result

# # Measure execution time on machine with 12 cores and 32G RAM
results_1 = []
for k in range(5,15):
    
    start_time_machine2 = time.time()
    result_machine2 = perform_task(k)
    end_time_machine2 = time.time()
    execution_time_machine2 = end_time_machine2 - start_time_machine2
    print("execution_time_machine2 :  ",execution_time_machine2)
    results_1.append(execution_time_machine2)

results_1
```

# time execution difference ploting

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3]  # X-axis values

results_1 = [125.246221,1600,16000]
    
results_2 = [45.72686147689819, 390.25975036621094,3882.2116572856903]


# Plot the lines
plt.plot(x, results_1, label='1 machine  --> RAM: 8Gb,  CPU: 3')
plt.plot(x, results_2, label='4 machines --> RAM: 32Gb, CPU: 12')

# Customize the plot
plt.xlabel('Iteration')
plt.ylabel('Execution time (S)')
plt.title('execution time difference ')
plt.legend()

# Show the plot
plt.show()

```
